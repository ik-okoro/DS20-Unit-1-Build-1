### Edit Data Frames

```
countries = [france, england, germany, italy, spain]
```

```
for country in countries:
  country["Date"] = pd.to_datetime(country["Date"], dayfirst= True, utc = False)
```

```
# Check all leagues were formatted correctly
# All leagues should start in August
for country in countries:
  print(country.Date[0].month)
```

```
# View column headers for editing
france.columns
```

```
# Combine leagues into one dataframe
merged = pd.concat([france, england, germany, italy, spain])
merged = merged.rename({"Division" : "League"}, axis = 1)
merged = merged.replace({"F1" : "France Ligue 1", "E0" : "England Premier League", "D1" : "Germany Bundesliga", "I1" : "Italy Serie A", "SP1" : "Spain La Liga"})
```

```
# Create function to identify each football season
import datetime

def season_maker(value):
    if value < datetime.date(2011, 6, 15):
        return "2010/2011 Season"
    if value > datetime.date(2011, 7, 1) and value < datetime.date(2012, 6, 15):
        return "2011/2012 Season"
    if value > datetime.date(2012, 7, 1) and value < datetime.date(2013, 6, 15):
        return "2012/2013 Season"     
    if value > datetime.date(2013, 7, 1) and value < datetime.date(2014, 6, 15):
        return "2013/2014 Season"  
    if value > datetime.date(2014, 7, 1) and value < datetime.date(2015, 6, 15):
        return "2014/2015 Season"  
    if value > datetime.date(2015, 7, 1) and value < datetime.date(2016, 6, 15):
        return "2015/2016 Season"  
    if value > datetime.date(2016, 7, 1) and value < datetime.date(2017, 6, 15):
        return "2016/2017 Season"  
    if value > datetime.date(2017, 7, 1) and value < datetime.date(2018, 6, 15):
        return "2017/2018 Season"
    if value > datetime.date(2018, 7, 1) and value < datetime.date(2019, 6, 15):
        return "2018/2019 Season"  
    if value > datetime.date(2019, 7, 1) and value < datetime.date(2020, 8, 31):
        return "2019/2020 Season"    
    else:
        return 'not in range'   
```

```
merged["Season"] = merged.Date.apply(season_maker)
```

```
merged.head()
```

```
# Create summary table for total stats between leagues
sum_pivot = pd.pivot_table(merged, index=["League", "Season"], values=["Total Goals", "Total Shots", "Total Fouls", "Total Shots on Target"], aggfunc='sum').reset_index()

sum_pivot
```

```
# Add a column for the efficiency of each league
sum_pivot["Accuracy"] = ((sum_pivot["Total Shots on Target"] / sum_pivot["Total Shots"])*100).round(2)
sum_pivot
```

```
# Create summary table for average stats per season between leagues
avg_pivot = pd.pivot_table(merged, index=["League", "Season"], values=["Total Goals", "Total Shots", "Total Fouls", "Total Shots on Target"], aggfunc='mean').round(2).reset_index()

avg_pivot
```

```
# Create a column for the average efficiency of each league through each season
avg_pivot["Efficiency"] = ((avg_pivot["Total Goals"] / avg_pivot["Total Shots"])*100).round(2)
avg_pivot
```
